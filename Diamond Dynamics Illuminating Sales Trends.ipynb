```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
diamonds_df = pd.read_csv('/kaggle/input/diamonds-sale-data/diamonds.csv')
```

```python
diamonds_df.head()
```

```python
diamonds_df.describe()
```

```python
diamonds_df.isnull().sum()
```

```python
# Distribution of diamond prices
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_df, x='price', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Diamond Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of carat vs price
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diamonds_df, x='carat', y='price', color='salmon')
plt.title('Carat vs Price')
plt.xlabel('Carat')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Boxplot of diamond cut vs price
plt.figure(figsize=(10, 6))
sns.boxplot(data=diamonds_df, x='cut', y='price', palette='pastel')
plt.title('Diamond Cut vs Price')
plt.xlabel('Cut')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Pairplot to visualize relationships between numerical variables
sns.pairplot(data=diamonds_df[['carat', 'depth', 'table', 'price']], diag_kind='kde')
plt.show()
```

```python
# Countplot of Diamond Cut
plt.figure(figsize=(8, 6))
sns.countplot(data=diamonds_df, x='cut', palette='Set2')
plt.title('Count of Diamonds by Cut')
plt.xlabel('Cut')
plt.ylabel('Count')
plt.show()
```

```python
# Boxplot of Diamond Color vs Price
plt.figure(figsize=(10, 6))
sns.boxplot(data=diamonds_df, x='color', y='price', palette='viridis')
plt.title('Diamond Color vs Price')
plt.xlabel('Color')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Violinplot of Diamond Clarity vs Price
plt.figure(figsize=(10, 6))
sns.violinplot(data=diamonds_df, x='clarity', y='price', palette='muted')
plt.title('Diamond Clarity vs Price')
plt.xlabel('Clarity')
plt.ylabel('Price ($)')
plt.show()
```

```python
# Distribution of Diamond Depth
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_df, x='depth', bins=30, kde=True, color='lightcoral')
plt.title('Distribution of Diamond Depth')
plt.xlabel('Depth')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pairplot of Carat, Depth, and Price colored by Diamond Cut
plt.figure(figsize=(10, 6))
sns.pairplot(data=diamonds_df[['carat', 'depth', 'price', 'cut']], hue='cut', diag_kind='kde')
plt.suptitle('Pairplot of Carat, Depth, and Price Colored by Cut', y=1.02)
plt.show()
```
